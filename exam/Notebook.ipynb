```python
strings = []
while True:
    user_input = input("Enter a string (press Enter to finish): ")
    if user_input == "":
        break
    strings.append(user_input)

strings_lower = [s.lower() for s in strings]
strings_lower.sort()
print("\nSorted strings in lowercase, alphabetical order:")
for s in strings_lower:
    print(s)
```
